1. Create a scatter plot to visualize the realationship between  two variable , by generating a synthetic dataset.  



```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generating a synthetic dataset
np.random.seed(42)  # For reproducibility
x = np.random.rand(100) * 10  # 100 random x values between 0 and 10
y = 2.5 * x + np.random.randn(100) * 5  # y is a linear function of x with some noise

# Creating a scatter plot to visualize the relationship
plt.scatter(x, y, color='blue', label='Synthetic Data')

# Adding titles and labels
plt.title('Scatter Plot of Synthetic Dataset')
plt.xlabel('x values')
plt.ylabel('y values')

# Adding a grid
plt.grid(True)

# Displaying the plot
plt.legend()
plt.show()
```

4. Generate a dataset with categories and numerical values. visualize the distribution of a numerical variable across different categories .  



```python
import numpy as np
import matplotlib.pyplot as plt

# Generating a dataset with categories and corresponding numerical values
categories = ['Category A', 'Category B', 'Category C', 'Category D']
# Creating random data for each category
values_A = np.random.normal(50, 10, 100)
values_B = np.random.normal(60, 15, 100)
values_C = np.random.normal(55, 20, 100)
values_D = np.random.normal(65, 5, 100)

# Combining all values into a list for visualization
data = [values_A, values_B, values_C, values_D]

# Creating a box plot to visualize the distribution of the numerical variable across categories
plt.boxplot(data, labels=categories, patch_artist=True)

# Adding titles and labels
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Categories')
plt.ylabel('Values')

# Displaying the plot
plt.show()
```

5. Generate a synthetic dataset with correlated features. visualize the correlation matrix of a dataset using a heatmap.  



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generating a synthetic dataset with correlated features
n_samples = 100
feature_1 = np.random.rand(n_samples) * 10
feature_2 = 2.5 * feature_1 + np.random.randn(n_samples) * 2  # correlated with feature_1
feature_3 = -1.2 * feature_1 + np.random.randn(n_samples) * 3  # negatively correlated with feature_1
feature_4 = np.random.rand(n_samples) * 10  # independent feature

# Creating a DataFrame to store the features
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Calculating the correlation matrix
correlation_matrix = data.corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Adding a title
plt.title('Correlation Matrix Heatmap')

# Displaying the heatmap
plt.show()
```
